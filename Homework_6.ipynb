```python
def encode(string, offset):
    '''this function takes arguments a string and an integer offset and returns the encoded cipher'''
    new_string = ''
    for i in string:
        if 'a' <= i <= 'z':
            new = ord(i) + offset
            if new > ord('z'):
                new = new - 26
            new_string += chr(new)
        elif 'A' <= i <= 'Z':
            new = ord(i) + offset
            if new > ord('Z'):
                new = new - 26
            new_string += chr(new)
        else:
            new = ord(i)
            new_string += chr(new)
    return new_string
```

```python
def decode(c, n):
    '''this function takes a argument c as a cipher and an integer n as offset and returns the decoded string'''
    new_string = ''
    for i in c:
        if 'a' <= i <= 'z':
            new = ord(i) - n
            if new < ord('a'):
                new = new + 26
            new_string += chr(new)
        elif 'A' <= i <= 'Z':
            new = ord(i) + n
            if new < ord('A'):
                new = new + 26
            new_string += chr(new)
        else:
            new = ord(i)
            new_string += chr(new)
    return new_string
```

```python
import string
def letter_count(string):
    '''the function takes a string and returns a tuple.
    The first element of the tuple is a dictionary,
    that contains every unique upper case of the letter
    and the percentage of time of the letter.
    The second element in the tuple is the number of letters of the string.'''
    dic = {'A': 0, 
           'B': 0, 
           'C': 0, 
           'D': 0, 
           'E': 0, 
           'F': 0, 
           'G': 0, 
           'H': 0, 
           'I': 0, 
           'J': 0,
           'K': 0, 
           'L': 0, 
           'M': 0, 
           'N': 0, 
           'O': 0, 
           'P': 0, 
           'Q': 0, 
           'R': 0, 
           'S': 0, 
           'T': 0, 
           'U': 0, 
           'V': 0, 
           'W': 0, 
           'X': 0, 
           'Y': 0, 
           'Z': 0}
    second_tuple = 0
    for i in string:
        if 'a' <= i <= 'z' or 'A' <= i <= 'Z':
            second_tuple += 1
            upper = i.upper()
            x = dic.get(upper)
            dic[upper] = x+1
    for k in dic:
        number = dic.get(k)
        dic[k] = number*100/second_tuple
    return (dic, second_tuple)
```
